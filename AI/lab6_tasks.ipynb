```python
#Task1 - Local Beam Search Modification
import heapq
graph = {
    'S': [('A', 3), ('B', 6), ('C', 5)],
    'A': [('D', 9), ('E', 8)],
    'B': [('F', 12), ('G', 14)],
    'C': [('H', 7)],
    'H': [('I', 5), ('J', 6)],
    'I': [('K', 1), ('L', 10), ('M', 2)],
    'J': [('N', 4)],
    'D': [],
    'E': [],
    'F': [],
    'G': [],
    'K': [],
    'L': [],
    'M': [],
    'N': []
}

def beam_search(start,goal,beam_width=2):
    goalFound = False
    while not goalFound and beam_width < 5:
        print(f"\nExploring with beam-width = {beam_width}")
        beam = [(0, [start])]  
        while beam:
            candidates = []
            for cost,path in beam:
                current_node = path[-1] 
                if current_node == goal:
                    return path,cost
                
                for neighbour,edge_cost in graph.get(current_node,[]):  
                    new_cost = cost+edge_cost   
                    new_path = path + [neighbour]
                    candidates.append((new_cost,new_path))

                beam = heapq.nsmallest(beam_width,candidates,key=lambda x:x[0])
        beam_width+=1

    return None, float('inf')

start = 'S'
goal = 'L'
k=2
path, cost = beam_search(start,goal,k)

if path:
    print(f"Path Found: {{{path}}} with cost={cost}")
else:
    print("Goal Not Found")
```

```python
#Task 2 - Hill Climbing with Random Restart (N-Queens)
import random
def calc_conflicts(state):
    print(state)
    conflicts=0
    n=len(state)    #state is 1D bcz index represent column
    for i in range(n):
        for j in range(i+1,n):
            if state[i]==state[j] or abs(state[i]-state[j])==abs(i-j):
                conflicts+=1

    return conflicts

def get_neighbours(state):
    neighbours = []
    n = len(state)
    for col in range(n):
        for row in range(n):
            if row != state[col]:   
                new_state = list(state)     
                new_state[col] = row
                neighbours.append(new_state)
    return neighbours

def simple_hill_climbing(n):
    current_state = []
    for _ in range(n):
        random_row = random.randint(0,n-1)
        current_state.append(random_row)
    current_conflicts = calc_conflicts(current_state)

    while True:
        neighbours = get_neighbours(current_state)
        next_state = None
        next_conflicts = current_conflicts

        for neighbour in neighbours:
            neighbour_conflicts = calc_conflicts(neighbour)
            if neighbour_conflicts < next_conflicts:
                next_state = neighbour
                next_conflicts = neighbour_conflicts
                break   

        if next_conflicts>=current_conflicts:
            break

        #move
        current_state=next_state
        current_conflicts=next_conflicts
    return current_state,current_conflicts
    

attempts=0
solution,conflicts = simple_hill_climbing(8)
while conflicts and attempts<20:
    print("Attempt", attempts+1)
    solution,conflicts = simple_hill_climbing(8)
    print("Stuck at Local Optima, Re-initializing with new random state")
    attempts+=1

if not conflicts:
    print(f"Solution found for N-Queen (after {attempts+1} Attempts):-\n")
    print(solution)
else:
    print("cant solve even after 20 restarts")
```

```python
#Task 3 - Genetic Algorithm to design class schedule
import random
numTeachers = 5
numCourses = 5
numSlotsPerDay = 5
numDaysPerWeek = 5
totalSlots = 5*5

courseScheduleCount = 3
maxConsecutiveClassToTeach = 3
populationSize = 20
maxGens = 200
mutationRate = 0.2

def createPopulation():
    #We create a 3d array like timetable[teacher][day][slot]
    schedule = [[[0 for _ in range(numSlotsPerDay)] for _ in range(numDaysPerWeek)] for _ in range(numTeachers)]
    
    allPossibleSlots = []
    for day in range(numDaysPerWeek):
        for slot in range(numSlotsPerDay):
            allPossibleSlots.append((day,slot))

    #for each teacher pick 3 random slots
    for teacher in range(numTeachers):
        assignedSlots = random.sample(allPossibleSlots,3)
        for day,slot in assignedSlots:
            schedule[teacher][day][slot] = 1

    return schedule

def evaluateFitness(schedule):
    #exact 3 classes per week is already solved in creating schedule
    penalty=0
    for teacher in range(numTeachers):
        for day in range(numDaysPerWeek):
            consecutiveSlots=0
            for slot in range(numSlotsPerDay):
                if schedule[teacher][day][slot]==1:
                    consecutiveSlots+=1
                else:
                    consecutiveSlots=0

                if consecutiveSlots>=3:
                    penalty+=10

    for day in range(numDaysPerWeek):
        for slot in range(numSlotsPerDay):
            collision=0
            for teacher in range(numTeachers):
                if schedule[teacher][day][slot]==1:
                    collision += 1
            if collision>1:
                penalty += (collision)*10

    return penalty

def selection(population,fitness_score):
    scoredPopulation = []
    for i in range(len(population)):
        pair = [fitness_score[i],population[i]]
        scoredPopulation.append(pair)
    
    scoredPopulation.sort()
    parent1 = scoredPopulation[0][1]
    parent2 = scoredPopulation[1][1]
    return parent1,parent2

def crossover(parent1,parent2):
    child = []
    for i in range(numTeachers):
        #flip a coin for each teacher and give it whole to the child
        if random.random() < 0.5:
            child.append(parent1[i])
        else:
            child.append(parent2[i])

    return child

def mutation(schedule):
    #pick 1 random teacher and swap its 2 (day,slot) pair  
    teacher = random.randint(0,numTeachers-1)

    day1 = random.randint(0,numDaysPerWeek-1)
    slot1 = random.randint(0,numSlotsPerDay-1)
    day2 = random.randint(0,numDaysPerWeek-1)
    slot2 = random.randint(0,numSlotsPerDay-1)

    schedule[teacher][day1][slot1], schedule[teacher][day2][slot2]  = schedule[teacher][day2][slot2], schedule[teacher][day1][slot1]
    return schedule

def geneticAlgorithm():
    population = [createPopulation() for _ in range(populationSize)]
    for generation in range(maxGens):
        fitness_scores = [evaluateFitness(schedule) for schedule in population]
        bestFitness = min(fitness_scores)
        print(f"Generation {generation+1} Best Fitness: {bestFitness}")

        parents = selection(population,fitness_scores)
        newPopulation = []

        while len(newPopulation) < populationSize:
            parent1,parent2 = random.sample(parents,2)
            child = crossover(parent1,parent2)
            if random.random() < mutationRate:
                child = mutation(child)

            newPopulation.append(child)
        population = newPopulation
    bestSchedule = population[fitness_scores.index(min(fitness_scores))]
    print("Best Schedule (Teachers x Days x Slots)")

    return bestSchedule

def print_timetable(best_schedule):
    days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
    print("\n--- MASTER SCHEDULE (Teacher ID in each Slot) ---")
    print("      Slot 1  Slot 2  Slot 3  Slot 4  Slot 5")
    print("      --------------------------------------")

    for d in range(numDaysPerWeek):
        row_display = []
        for s in range(numSlotsPerDay):
            teachers_in_slot = []
            for t in range(numTeachers):
                if best_schedule[t][d][s] == 1:
                    teachers_in_slot.append(str(t))
            
            # If no teacher, show ".", if one teacher show their ID, if collision show "!"
            if not teachers_in_slot:
                row_display.append("  .   ")
            elif len(teachers_in_slot) == 1:
                row_display.append(f"  T{teachers_in_slot[0]}  ")
            else:
                # This helps you spot where the GA failed to resolve a collision
                row_display.append(f" !{'+'.join(teachers_in_slot)}! ")

        print(f"{days[d]}: " + "".join(row_display))
    print("--------------------------------------")

schedule = geneticAlgorithm()
print_timetable(schedule)
```
