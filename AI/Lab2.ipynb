```python
#Task1
class Student:
    def __init__(self,name,roll_no,program):
        self.name = name
        self.roll_no = roll_no
        self.program = program

    def display_info(self):
        print(f"Name = {self.name}")
        print(f"Roll = {self.roll_no}")
        print(f"Program = {self.program}\n")

def main():
    s1 = Student("Syed Sultan Ahmed","24K-0585","BS-CS")
    s2 = Student("Syed Ali Hossein","24K-0578","Media Science")
    s1.display_info()
    print("-" * 49)
    s2.display_info()
main()
```

```python
#Task2
class SmartFan_Controller:
    def __init__(self,room_name,speed):
        self.room_name = room_name
        self.speed = speed

    def increase_speed(self,s):
        self.speed += s

    def decrease_speed(self,s):
        self.speed -= s
        if self.speed < 0:
            speed = 0

    def display_status(self):
        print(f"Room Name = {self.room_name}")
        print(f"Speed = {self.speed}\n")

def main():
    f1 = SmartFan_Controller("R-108",5)
    f2 = SmartFan_Controller("E-28",3)
    f1.display_status()
    f2.display_status()
    f1.increase_speed(3)
    f2.decrease_speed(2)
    print("-" * 49)
    f1.display_status()
    f2.display_status()

main()
```

```python
#Task4
class BankAccount:
    def __init__(self,account_holder,account_number,balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self,amount):
        self.balance += amount

    def withdraw(self,amount):
        self.balance -= amount
        if self.balance < 0:
            self.balance = 0

    def display_account(self):
        print(f"Account Holder = {self.account_holder}")
        print(f"Account Number = {self.account_number}")
        print(f"Balance = {self.balance}\n")

def main():
    acc1 = BankAccount("Syed Sultan",123456,5000)
    acc2 = BankAccount("Syed Khizer", 987654,6301)
    acc1.deposit(500)
    acc2.withdraw(1000)
    acc1.display_account()
    print("-"*49)
    acc2.display_account()
main()
```

```python
#Task5
class SmartLight:
    def __init__(self,room,status):
        self.room = room
        self.status = status

    def turn_on(self):
        self.status = 1
    def turn_off(self):
        self.status = 0
    def show_status(self):
        print("Light of", self.room ," is ", end='')
        print("off" if self.status == 0 else "onn")

def main():
    l1 = SmartLight("D-28",0)
    l2 = SmartLight("Sports Room",1)
    l1.turn_on()
    l2.turn_off()
    l1.show_status()
    l2.show_status()
main()
```

```python
#Task6
class Staff:
    def __init__(self,name,staff_id,department):
        self.name = name
        self.staff_id = staff_id
        self.department = department

    def display_info(self):
        print(f"Name = {self.name}")
        print(f"Staff ID = {self.staff_id}")
        print(f"Department= {self.department}")

class Teacher(Staff):
    def __init__(self,name,staff_id,department,courses,salary):
        super().__init__(name,staff_id,department)
        self.courses = courses
        self.salary = salary

    def display_info(self):
        super().display_info()
        print(f"Courses = {self.courses}")
        print(f"Salary = {self.salary}\n")

class AdminStaff(Staff):
    def __init__(self,name,staff_id,department,role,working_hours):
        super().__init__(name,staff_id,department)
        self.role = role
        self.working_hours = working_hours

    def display_info(self):
        super().display_info()
        print(f"Role = {self.role}")
        print(f"Working hours = {self.working_hours}\n")

class ResearchAssistant(Staff):
    def __init__(self,name,staff_id,department,research_topic,stipend):
        super().__init__(name,staff_id,department)
        self.research_topic = research_topic
        self.stipend = stipend

    def display_info(self):
        super().display_info()
        print(f"Research topic = {self.research_topic}")
        print(f"Stipend = {self.stipend}\n")

def main():
    t1 = Teacher("Riaz Ahmed","#01234","AI","{AI,DL,AI,ML,NLP}", 230000)
    t2 = Teacher("Abdullah Shaikh","#04321","EE","{OS,COAL,CA,DLD}", 200000)
    a1 = AdminStaff("Talha Shahid","#00123","Finance","Accountant",90000)
    a2 = AdminStaff("Murtaza Liaquat","#00321","Mantainence", "Technician",60000)
    r1 = ResearchAssistant("Hunain Ali", "#00012", "AI", "AGI", 25000)
    r2 = ResearchAssistant("Mubin Farid", "#00021", "CS", "Quantum computing", 35000)

    t1.display_info()
    t2.display_info()
    print("*"*49)
    a1.display_info()
    a2.display_info()
    print("*"*49)
    r1.display_info()
    r2.display_info()
main()
```

```python
#Task7
class Vehicle:
    def __init__(self):
        pass

    def start(self):
        print("Engine Ignited")

class Car(Vehicle):
    def __init__(self):
        pass

    def start(self):
        print("Car Started")

class Bike(Vehicle):
    def __init__(self):
        pass

    def start(self):
        print("Bike Started")

class Bus(Vehicle):
    def __init__(self):
        pass

    def start(self):
        print("Bus Started")

def main():
    v = Vehicle()
    c = Car()
    b = Bike()
    bu = Bus()

    v.start()
    c.start()
    b.start()
    bu.start()
main()
```

```python
#Task8
class Calculator:
    def __init__(self):
        pass

    def multiply(self,n1,n2):
        return n1*n2
        
    def multiply(self,n1,n2,n3):
        return n1*n2*n3

    def multiply(self,*args):
        result = 1
        for n in args:
            result *= n
        return result

def main():
    c = Calculator()
    print(c.multiply(2,3))
    print(c.multiply(5,2,3))
    print(c.multiply(1,2,3,4,5))
main()
```

```python
#Task9
class Employee:
    def __init__(self,name,salary):
        self._name = name
        self._salary = salary

    def get_name(self):
        return self._name
    def get_salary(self):
        return self._salary

class Manager(Employee):
    def __init__(self,name,age):
        super().__init__(name,age)

    def show_employee_details(self):
        print(f"Employee Name = {self._name}")
        print(f"Employee Salary = {self._salary}")

def main():
    m = Manager("Syed Sultan",250000)
    m.show_employee_details()
main()
```
