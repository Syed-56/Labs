```python
students = [
        {"name":"Sultan", "marks":86, "grade":''},
        {"name":"Hammad", "marks":75, "grade":''},
        {"name":"Ali H", "marks":66, "grade":''},
        {"name":"Atif", "marks":56, "grade":''}
    ]

for s in students:
    if s["marks"] > 86:
        s["grade"] = 'A'
    elif s["marks"] > 76:
        s["grade"] = 'B'
    elif s["marks"] > 66:
        s["grade"] = 'C'
    elif s["marks"] > 56:
        s["grade"] = 'D'
    else:
        s["grade"] = 'F'

for s in students:
    print(s)
```

```python
num = int(input("Enter: "))
isPrime = False
for i in range(int(num**0.5)):
    if i % 2 != 0:
        print("Prime")
        isPrime = True
        
if not isPrime:
    print("Not Prime") 
```

```python
def get_max_number(*args):
    return max(args)

def print_student_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print(get_max_number(10, 55, 30, 99, 4))

print_student_info(name="Sultan", age=22, grade="A")
```
