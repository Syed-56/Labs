```python
#Task1
tree = {
    'A': [['B', 2], ['C', 5]],
    'B': [['D', 1], ['E', 7]],
    'C': [['F', 3], ['G', 2]],
    'D': [['H', 4], ['I', 8]],
    'E': [['J', 1], ['K', 3]],
    'F': [['L', 6], ['M', 2]],
    'G': [['N', 5], ['O', 1]],
    'H': [['P', 3], ['Q', 4]],
    'I': [['R', 2], ['S', 9]],
    'J': [['T', 1], ['U', 2]],
    'K': [['V', 5], ['W', 3]],
    'L': [['X', 1], ['Y', 4]],
    'M': [['Z', 2]], # Z is the 26th node
    'N': [], 'O': [], 'P': [], 'Q': [], 'R': [], 
    'S': [], 'T': [], 'U': [], 'V': [], 'W': [], 
    'X': [], 'Y': [], 'Z': []
}

def bfs(tree,start,goal):
    print("BFS = ",end="")
    visited = []
    queue = []
    visited.append(start)
    queue.append(start)

    while queue:
        node = queue.pop(0)
        if node == goal:
            print(node, "  ----Goal Found")
            break
        
        print(node+"->",end="")
        for neighbour in tree[node]:
            name = neighbour[0]
            if name not in visited:
                visited.append(name)
                queue.append(name)

def dfs(tree,start,goal):
    print("DFS = ",end="")
    visited = []
    stack = []
    visited.append(start)
    stack.append(start)

    while stack:
        node = stack.pop()
        if node == goal:
            print(node, "\nGoal Found")
            break

        print(node+"->", end="")
        for neighbour in reversed(tree[node]):
            name = neighbour[0]
            if name not in visited:
                visited.append(name)
                stack.append(name)

def dls(tree,start,goal,limit):
    visited = []
    stack = []
    depth_stack = [0]
    visited.append(start)
    stack.append(start)

    while stack:
        node = stack.pop()
        depth = depth_stack.pop()

        if node == goal:
            print(node+"\n Goal Found")
            return True
        if depth > limit:
            print("------Goal Not Found")
            return False
        else:
            print(node+"->",end="")
            visited.append(node)

            for neighbour in reversed(tree[node]):
                name = neighbour[0]
                if name not in visited:
                    stack.append(name)
                    depth_stack.append(depth+1)

def ids(tree,start,goal,max_depth=5):
    print("\n\nIDS:-")
    for i in range(max_depth):
        found = dls(tree,start,goal,i)
        print("")
        if found:
            print("IDS found goal at depth: " + str(i))
            return True

    print("Goal not found within max depth " + str(max_depth))
    return False

def ucs(tree,start,goal):
    print("\nUCS:-")
    frontier = []
    visited = []
    frontier.append((start,0))
    cost_soFar = {start:0}
    came_from = {start:None}    #for path reconstruction

    while frontier:
        frontier.sort(key=lambda x: x[1])   #sort to simulate PQ
        currentNode, currentCost = frontier.pop(0)
        if currentNode in visited:
            continue
        
        visited.append(currentNode)
        if currentNode == goal:
            path = []
            while currentNode is not None:
                path.append(currentNode)
                currentNode = came_from[currentNode]
            path.reverse()
            print(f"{path} with cost = {currentCost}")
            return

        for neighbour, cost in tree[currentNode]:
            newCost = currentCost + cost
            if neighbour not in cost_soFar or newCost < cost_soFar[neighbour]:
                cost_soFar[neighbour] = newCost
                came_from[neighbour] = currentNode
                frontier.append((neighbour,newCost))
    print("Goal Not Found")

def main():
    start_node = 'A'
    goal_node = 'G'
    bfs(tree,start_node,goal_node)
    dfs(tree,start_node,goal_node)
    print("DLS = ", end="")
    dls(tree,start_node,goal_node,2)
    ids(tree,start_node,goal_node)
    ucs(tree,start_node,goal_node)

main()
```

```python

# Task2
romania_map = {
    'Arad': [['Zerind', 75], ['Sibiu', 140], ['Timisoara', 118]],
    'Zerind': [['Arad', 75], ['Oradea', 71]],
    'Oradea': [['Zerind', 71], ['Sibiu', 151]],
    'Sibiu': [['Arad', 140], ['Oradea', 151], ['Fagaras', 99], ['Rimnicu Vilcea', 80]],
    'Timisoara': [['Arad', 118], ['Lugoj', 111]],
    'Lugoj': [['Timisoara', 111], ['Mehadia', 70]],
    'Mehadia': [['Lugoj', 70], ['Drobeta', 75]],
    'Drobeta': [['Mehadia', 75], ['Craiova', 120]],
    'Craiova': [['Drobeta', 120], ['Rimnicu Vilcea', 146], ['Pitesti', 138]],
    'Rimnicu Vilcea': [['Sibiu', 80], ['Craiova', 146], ['Pitesti', 97]],
    'Fagaras': [['Sibiu', 99], ['Bucharest', 211]],
    'Pitesti': [['Rimnicu Vilcea', 97], ['Craiova', 138], ['Bucharest', 101]],
    'Bucharest': [['Fagaras', 211], ['Pitesti', 101], ['Urziceni', 85], ['Giurgiu', 90]],
    'Giurgiu': [['Bucharest', 90]],
    'Urziceni': [['Bucharest', 85], ['Vaslui', 142], ['Hirsova', 98]],
    'Hirsova': [['Urziceni', 98], ['Eforie', 86]],
    'Eforie': [['Hirsova', 86]],
    'Vaslui': [['Urziceni', 142], ['Iasi', 92]],
    'Iasi': [['Vaslui', 92], ['Neamt', 87]],
    'Neamt': [['Iasi', 87]]
}

def ucs_romania(graph, start, goal):
    frontier = [[start, 0]]
    visited = {} 
    came_from = {start: None}
    
    while frontier:
        frontier.sort(key=lambda x: x[1])
        current_node, current_cost = frontier.pop(0)
        
        if current_node == goal:
            path = []
            temp = current_node
            while temp is not None:
                path.append(temp)
                temp = came_from[temp]
            path.reverse()
            return path, current_cost
            
        if current_node in visited and visited[current_node] <= current_cost:
            continue
            
        visited[current_node] = current_cost
        
        for neighbor, weight in graph[current_node]:
            new_cost = current_cost + weight
            if neighbor not in visited or new_cost < visited[neighbor]:
                came_from[neighbor] = current_node
                frontier.append([neighbor, new_cost])

path, total_cost = ucs_romania(romania_map, 'Arad', 'Bucharest')

print("Path from Arad to Bucharest:")
print(" -> ".join(path))
print("Total Distance: " + str(total_cost))
```

```python
class PuzzleEnvironment:
    def __init__(self, start, goal):
        self.state = start
        self.goal = goal

    def get_neighbors(self, state):
        neighbors = []
        idx = state.index(0)
        row, col = divmod(idx, 3)
        
        moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]
        for dr, dc in moves:
            r, c = row + dr, col + dc
            if 0 <= r < 3 and 0 <= c < 3:
                new_idx = r * 3 + c
                new_state = list(state)
                new_state[idx], new_state[new_idx] = new_state[new_idx], new_state[idx]
                neighbors.append(tuple(new_state))
        return neighbors

class GoalBasedAgent:
    def __init__(self, goal):
        self.goal = goal

    def bfs_solve(self, env):
        start = env.state
        visited = set() 
        queue = []
        
        visited.add(start)
        queue.append((start, []))

        count = 0
        while queue:
            node, path = queue.pop(0)
            count += 1
            
            # Print progress every 10000 nodes so you know it's working
            if count % 10000 == 0:
                print("Checking state number: " + str(count))

            if node == self.goal:
                return path + [node]

            for neighbour in env.get_neighbors(node):
                if neighbour not in visited:
                    visited.add(neighbour)
                    queue.append((neighbour, path + [node]))
        return None

def run_agent():
    start_state = (7, 2, 4, 5, 0, 6, 8, 3, 1)
    goal_state = (0, 1, 2, 3, 4, 5, 6, 7, 8)

    env = PuzzleEnvironment(start_state, goal_state)
    agent = GoalBasedAgent(goal_state)

    result_path = agent.bfs_solve(env)

    if result_path:
        print("\n--- Goal Reached! ---")
        final_state = result_path[-1]
        for i in range(0, 9, 3):
            print(final_state[i:i+3])
        print("Total Moves: " + str(len(result_path) - 1))

run_agent()
```
